```python

df = pd.read_csv('Google_stock.csv')  
print(df.head())
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['5-Day MA'] = [df.iloc[i-4:i+1]['Close'].mean() if i >= 4 else None for i in range(len(df))]
df['10-Day MA'] = [df.iloc[i-9:i+1]['Close'].mean() if i >= 9 else None for i in range(len(df))]
print(df.iloc[:, [0, 1, -2, -1]].head()) 
df.to_csv('sample2_with_moving_averages.csv', index=False)
```

```python

df_2024 = pd.read_csv('Google_stock.csv', parse_dates=['Date']).loc[
    (pd.read_csv('Google_stock.csv', parse_dates=['Date'])['Date'].dt.year == 2024)
].head(100)
df_2024['5-Day MA'] = df_2024['Close'].rolling(5).mean()
df_2024['10-Day MA'] = df_2024['Close'].rolling(10).mean()

# Plot the data
df_2024.plot(x='Date', y=['Close', '5-Day MA', '10-Day MA'], figsize=(10, 6), 
             title='Stock Price and Moving Averages for 2024 (First 100 Rows)', 
             ylabel='Price', xlabel='Date', style=['-', '--', '--'], 
             color=['blue', 'red', 'green'])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
